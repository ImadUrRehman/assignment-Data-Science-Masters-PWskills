Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that allows users to create a wide range of high-quality 2D and 3D plots, charts, and graphs. It is widely used by data analysts, scientists, and researchers to visualize their data in a clear and concise manner.

Line plots: A line plot is a graph that displays data as points connected by straight lines. It is used to visualize trends in data over time.

Bar plots: A bar plot is a graph that represents categorical data with rectangular bars. It is used to compare data across categories.

Scatter plots: A scatter plot is a graph that displays the relationship between two variables as a collection of points. It is used to visualize the correlation between variables.

Histograms: A histogram is a graph that represents the distribution of numerical data by dividing it into bins and displaying the frequency of observations in each bin.

Pie charts: A pie chart is a circular graph that displays data as slices of a pie. It is used to represent proportions and percentages.

Q2: What is a scatter plot?

A scatter plot is a type of plot or graph that displays the relationship between two numerical variables
In a scatter plot, each point on the plot represents a pair of values from the two variables being compared. One variable is plotted along the x-axis, and the other variable is plotted along the y-axis. Each point's position on the graph is determined by its corresponding values on the two axes.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,2))
plt.scatter(x,y)
plt.xlabel("This is my X")
plt.ylabel("This is my Y")
plt.title("this is my scatter plot")
```

Q3: Why is the subplot() function used?

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide the figure into a grid of rows and columns and specify the position of each plot within that grid.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x1,y1)
ax.plot(x2,y2)
ax.plot(x3,y3)
ax.plot(x4,y4)
```

Q4: What is a bar plot?

A bar plot, also known as a bar chart, is a graphical representation of data that uses rectangular bars to compare the values of different categories or groups. In a bar plot, the height or length of each bar represents the magnitude of the data being displayed. The bars are typically arranged horizontally or vertically, with one axis representing the categories or groups being compared and the other axis representing the magnitude or frequency of the data. Bar plots can be used to visualize a wide range of data types, including numerical data, categorical data, and time-series data, and can be created using various software tools such as Excel, Python, and R. They are commonly used in fields such as statistics, economics, business, and social sciences to present data in a clear and concise manner.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color='r')
plt.ylabel("PROFIT")
plt.xlabel("COMPANIES")
plt.title("BAR")
```

```python
plt.barh(company,profit)
plt.xlabel("PROFIT")
plt.ylabel("COMPANIES")
plt.title("H BAR")
```

Q5: What is a box plot? Why is it used?

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data along with its median, quartiles, and any outliers.

A box plot is typically made up of a box, which represents the middle 50% of the data, and two whiskers, which extend from the box to represent the range of the data. The box itself is defined by its lower and upper quartiles, which represent the middle 50% of the data, and its median, which represents the midpoint of the data. The whiskers extend from the box to the smallest and largest non-outlier data points, or a fixed distance from the box, depending on the specific rules used to create the box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig = plt.figure()
ax = fig.add_subplot()
ax.boxplot(box1)
ax.boxplot(box2)
```
