### Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask is a lightweight and flexible web application framework for Python. It is designed to make getting started with web development quick and easy, with minimalistic and simple coding requirements. Flask is not opinionated and allows developers to have more control over their web application's structure and design.

Some of the advantages of using Flask include:

Lightweight: Flask is a lightweight framework and does not require a lot of setup or external libraries.

Flexibility: Flask is very flexible and allows developers to choose their own tools and libraries to use in their application.

Easy to use: Flask has a simple and intuitive interface which makes it easy for developers to learn and use.

Good for small to medium-sized projects: Flask is well-suited for building small to medium-sized web applications.

Flask extensions: Flask has a large number of extensions available which can be used to add additional functionality to your application, such as user authentication, database integration, and more.

Scalability: Flask is a scalable framework and can be used to build applications of all sizes.

Community support: Flask has a large and active community of developers who contribute to the framework, provide support, and create useful extensions and libraries.

### Q3. What is App routing in Flask? Why do we use app routes?
In Flask, App routing refers to the process of defining routes for the application. A route is a URL pattern that Flask uses to match incoming requests to the appropriate view function. When a user sends a request to a Flask application, the application uses the route defined to determine which function should handle the request.

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint and arguments based on the routing information provided.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))  # Output: /
        print(url_for('profile', username='Alice'))  # Output: /user/Alice
```
